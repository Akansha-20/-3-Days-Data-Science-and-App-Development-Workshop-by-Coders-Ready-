```python
import geopandas as gpd
```

```python
gdf = gpd.read_file('/content/sample_data/shp')
gdf
```

```python
h = gdf[gdf['NAME_1']=='Haryana'].reset_index(drop= True)
h
```

```python
hd= list(h['NAME_2'])
```

```python
hd
```

```python
h_div = h.loc[h['NAME_2'].isin(hd)]
```

```python
h_div.reset_index(drop=True , inplace=True)
h_div
```

```python
import rasterio
import rasterio.features

geometry=h_div['geometry']

mask= rasterio.features.geometry_mask(geometry, out_shape = xds.shape[-2:], transform = xds.rio.transform(), invert= True)

masked_xds = xds.where(mask)
```

```python
masked_xds.shape
```

```python
from click.formatting import iter_rows
fig, ax = plt.subplots(figsize=(8,8))

img = masked_xds[0].plot.imshow(vmin=0, vmax=1500, cmap='terrain', ax=ax, cbar_kwargs={
    'orientation' : 'vertical',
    'label' : 'Mean Sea level Height (in m)',
})

ax.set_xlabel = ('Longitudes due East')
ax.set_ylabel = ('latitude due North')

h_div.plot(ec='k', fc = 'none', alpha = 0.4, marker='--',ax=ax)

x,y = (76.768688, 28.450978)
plt.plot(x,y ,'ok', markersize=5)
plt.text(x+0.04 ,y+0.01, 'my college ', fontsize=10)

for idx, row in h_div.iterrows():
  centroid= row['geometry'].centroid
  ax.text(centroid.x+0.07, centroid.y+0.07, row['NAME_2'], ha= 'center', va= 'center', fontsize=6, color='grey')

ax.set_title('')
ax.set_aspect('auto')
plt.tight_layout()
plt.savefig('mycollege.png', dpi=600 ,bbox_inches='tight')
plt.show()
```
